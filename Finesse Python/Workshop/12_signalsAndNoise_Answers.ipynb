### Task 1: Loop transfer functions

As we did for the fields in the Michelson Interometer and Fabry-PÃ©rot cavity examples, we can evaluate the *steady-state*, point-to-point transfer functions for the above block diagram. These are important in evaluating the relative importance of various noise sources in the interferometer.

1. Adopting the notation shown in the graphic above, evaluate the following transfer functions in terms of the Laplace-domain functions $P$ (= plant), and $C$ (= controller):
    - $H_{yr}$
    - $H_{en}$
    - $H_{ed}$
    - $H_{er}$
    
<span style="color:green"> **Answer**:  </span>
    
<span style="color:green">$H_{yr} = \frac{PC}{1+PC}$</span>,
<span style="color:green">$H_{en} = \frac{-1}{1+PC}$</span>,
<span style="color:green">$H_{ed} = \frac{-P}{1+PC}$</span>,
<span style="color:green">$H_{er} = \frac{1}{1+PC}$</span>


The first letter in the subscript is the measurement point, while the second is the source point. In this abstract representation, the units of this transfer function are dimensionless. Sometimes, the product $PC \equiv L$ is assigned a special name, called the "Open-loop gain". 

2. Qualitatively, what is the behaviour of these transfer functions in the regimes $L \gg 1$ (sometimes called in-loop) and $L \ll 1$ (sometimes called out-of-loop)?

<span style="color:green"> **Answer**:  </span>

<span style="color:green">When $L \gg 1$, the denominator of the above transfer functions is large. So, the transfer function from sensor noise, $n$, to the error signal, $e$, denoted as $|H_{en}|$, is small - in this regime, the sensor adds only a negligible amount of noise to the system as a result of closing the feedback loop. $|H_{yr}| \approx 1$, which means that when the feedback loop is closed, the system output $y$ is the same as the reference, $r$, which is the desired behavior.</span>

<span style="color:green">In the converse regime, $L \ll 1$, the denominator of the above transfer functions is small. Then, $|H_{en}| \approx 1$ - that is, the sensor noise is imprinted on the system output with almost no attenuation. $|H_{yr}| \ll 1$, which means that when the feedback loop is closed, the system output $y$ does not track the reference input well. In this regime, the feedback loop does not improve the stability of the system.</span>

### Noise sources budgets

There are many noise sources that limit the sensitivity of our interferometers. There are many ways to define signal and noise, but for our purposes:
 - Signal = phase modulation of the measured photocurrent by gravitational waves.
 - Noise = phase modulation of the measured photocurrent by anything else.

```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sig
```

```python
# Matplotlib setup
plt.style.use('fivethirtyeight')
plt.rcParams['lines.linewidth'] = 3
plt.rcParams['font.size'] = 22
plt.rcParams['axes.titlesize']  = 'large'
plt.rcParams['axes.labelsize']  = 'large'
plt.rcParams['xtick.labelsize'] = 'large'
plt.rcParams['ytick.labelsize'] = 'large'
plt.rcParams['axes.formatter.limits'] = [-2,2]
plt.rcParams['axes.grid'] = True
plt.rcParams['grid.linestyle'] = '--'
plt.rcParams['grid.linewidth'] = 0.7
plt.rcParams['grid.alpha'] = 0.4
plt.rcParams['text.usetex'] = False
plt.rcParams['legend.loc'] = 'best'
plt.rcParams['legend.fontsize'] = 'small'
plt.rcParams['figure.figsize'] = [12,9]
plt.rcParams['figure.facecolor'] = 'white'
plt.rcParams['figure.subplot.left'] = 0.07
plt.rcParams['figure.subplot.right'] = 0.95
plt.rcParams['figure.subplot.top'] = 0.92
```

```python
# Helper functions
def rms_timeDomain(V):
    '''
    Time domain RMS.
    
    Parameters:
    ------------
    V: array_like
        Time domain signal to compute RMS for
    
    Returns:
    ------------
    Vrms: float
        The RMS of the signal V.
    '''
    Vsq = V**2
    return(np.sqrt(np.mean(Vsq)))

def cum_rms(ff, asd):
    '''
    Cumulative RMS for an ASD.
    
    Parameters:
    -------------
    ff: array_like
        Frequency vector 
    asd: array_like
        ASD vector
        
    Returns:
    ---------
    cum_rms: array_like
        The cumulative RMS as a function of frequency. 
        The -1 element is the total RMS.
    '''
    df = np.roll(ff, -1) - ff
    df = df[:-1]
    cum_ms = np.zeros(len(ff))
    for ii in range(len(cum_ms)):
        cum_ms[ii] = np.trapz(asd[ii:]**2, ff[ii:])
    return(cum_ms**0.5)
```

```python
fs = 100                            # Sampling frequency [Hz]
tDur = 10                           # Total length of signal [s] 
nSamples = 10 / (1/fs) + 1          # Number of samples
# print (int(nSamples))
t = np.linspace(0,tDur,nSamples)    # Time vector [s]
# print (t)
Vn = np.random.randn(int(nSamples)) # Noise vector [a.u.]
# print (Vn)
Vrms = rms_timeDomain(Vn)           # RMS noise value [a.u.]
print('The RMS value of the signal is {} in arbitrary units. Is this what you expect given how the noise was generated?'.format(round(Vrms,3)))
```

```python
# FFT parameters
t_fft = 1                                   # FFT segment length [s]
nFFT = int(t_fft * fs)                      # Number of points in FFT segment
win = sig.get_window(('tukey',0.25), nFFT)  # FFT window
```

```python
# Do the FFT
ff, Pxx = sig.welch(Vn, fs=fs, window=win, 
                    scaling='density')
rms_f = cum_rms(ff, np.sqrt(Pxx))
```

```python
# Make a plot
fig, ax = plt.subplots(1,1,figsize=(16,9))
ax.loglog(ff, np.sqrt(Pxx), label='ASD', 
          color='xkcd:bright blue',
         rasterized=True, linewidth=3, alpha=0.7)
ax.loglog(ff, rms_f, linestyle='--', 
          color='xkcd:bright blue', alpha=0.7,
         rasterized=True, linewidth=3, label='RMS')
ax.legend(loc='best')
ax.set_xlabel('Frequency [Hz]')
ax.set_ylabel('ASD [$\\frac{\\mathrm{a.u.}}{\\sqrt{\\mathrm{Hz}}}$]')
fig.suptitle('ASD of simulated Gaussian white noise');
print('The RMS value computed from the frequency domain is {} a.u.'.format(round(rms_f[0],3)))
print('That from the time domain calculation was {} a.u.'.format(round(Vrms,3)))
```
