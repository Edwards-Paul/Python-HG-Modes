```python
import PauLisa as pl
```

```python
a = pl.defaultParams
b = pl.defaultPlane
```

```python
# (0,0), (1,0), and (0,0)+(0,1) modes NEAR
# Re[a10]/a00 = delta s/wo , w0 = 1e-3
# Im[a20]/a00 = (delta s)*Zr

m0 = pl.Modes((0,0,1))
m1 = pl.Modes((1,0,1))
mm = pl.Modes((1,0,.1*1j),(0,0,1))

f0 =pl.Calculate(a,b,m0,8.5)
faa = pl.Calculate(a,b,m0,3)
fbb = pl.Calculate(a,b,m0,5)
fcc = pl.Calculate(a,b,m0,0)
f1 = pl.Calculate(a,b,m1,1.5)
fmm = pl.Calculate(a,b,mm,1.5)


pl.AmplitudeSliceX(0,f0,f1,fmm)

pl.IntensitySliceX2(0,f0,faa,fbb,fcc,xlim=[-1e-3,1e-3])

pl.PhaseContour(f1,xlim=[-1e-3,1e-3],ylim=[-1e-3,1e-3])
pl.PhaseSliceX(0,f1,xlim=[-1e-3,1e-3])
```
