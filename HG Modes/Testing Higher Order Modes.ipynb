```python
import PauLisa as pl
```

```python
a = pl.defaultParams
b = pl.defaultPlane
```

```python
# (0,0), (1,0), and (0,0)+(0,1) modes NEAR
m0 = pl.Modes((0,0,1))
m1a = pl.Modes((1,0,1))
m1b = pl.Modes((1,0,1))
mm = pl.Modes((1,0,0.5),(0,0,1))

f0 =pl.Calculate(a,b,m0,1)
f1 = pl.Calculate(a,b,m1b,1)
fmm = pl.Calculate(a,b,mm,1)

pl.AmplitudeSliceX(0,f0,f1,fmm, xlim=[-.002,.002])


pl.AmplitudeSliceX(0,f0,f1,fmm, xlim=[-.001,.001])
pl.IntensitySliceX(0,f0,f1,fmm, xlim=[-.001,.001])
```

```python
f0 =pl.Calculate(a,b,m0,10)
f1 = pl.Calculate(a,b,m1b,10)
fmm = pl.Calculate(a,b,mm,10)

pl.AmplitudeSliceX(0,f0,f1,fmm, xlim=[-.005,.005])
pl.IntensitySliceX(0,f0,f1,fmm, xlim=[-.001,.001])
```
