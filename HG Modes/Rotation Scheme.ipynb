```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,11)
z = np.linspace(0,10,11)
zero = [0]*11
```

# Shift + Rotation with Initial Coordinates (x=0)

## About RX Beam

\begin{equation}
x' = a+x\cos\alpha + z \sin \alpha
\end{equation}

\begin{equation}
z' = z\cos\alpha - x \sin\alpha
\end{equation}

```python
a = 1
c = .2
one = [a]*11
two = [-a]*11
b = np.pi/4

def x2(z,x,a,b):
    return( a+(x)*np.cos(b)+z*np.sin(b))

def z2(x,z,a,b):
    return(z*np.cos(b)-(x)*np.sin(b))
```

```python
plt.figure(figsize=(9,6))

plt.plot(x,zero, label = "LO HG00")
plt.plot(x,one, label = "RX Tophat")

plt.plot(z2(0,z,a,-np.pi/3),x2(z,0,a,-np.pi/3), label='-pi/3')
plt.plot(z2(0,z,a,np.pi/3),x2(z,0,a,np.pi/3), label='pi/3')
plt.plot(z2(0,z,a,np.pi/6),x2(z,0,a,np.pi/6), label='pi/6')
plt.plot(z2(0,z,a,-np.pi/6),x2(z,0,a,-np.pi/6), label='-pi/6')
plt.plot(z2(0,z,a,np.pi/2),x2(z,0,a,np.pi/2),label='pi/2')
plt.plot(z2(0,z,a,-np.pi),x2(z,0,a,-np.pi),label='pi')


plt.legend()
plt.xlabel('z')
plt.ylabel('x')
plt.grid()
```

## About LO Beam

\begin{equation}
x' = (x+a)\cos\alpha + z \sin \alpha
\end{equation}

\begin{equation}
z' = z\cos\alpha - (x+a) \sin\alpha
\end{equation}

```python
a = 1.0
one = [a]*11
two = [-a]*11
b = np.pi/4
x_cor=0

def x2(z,x,a,b):
    return( (x+a)*np.cos(b)+z*np.sin(b)  )

def z2(x,z,a,b):
    return(z*np.cos(b)-(x+a)*np.sin(b) )
```

# Transforming Coordinates (x = x+a)

## About RX

\begin{equation}
x' = (x-a)\cos\alpha + z \sin \alpha = z \sin \alpha
\end{equation}

\begin{equation}
z' = z\cos\alpha - (x-a) \sin\alpha = z\cos\alpha 
\end{equation}

```python
a = 1.0
one = [a]*11
two = [-a]*11
b = np.pi/4
x_cor=0

def x2(z,x,a,b):
    return( (x-a)*np.cos(b)+z*np.sin(b)  )

def z2(x,z,a,b):
    return(z*np.cos(b)-(x-a)*np.sin(b) )
```

```python
plt.figure(figsize=(9,6))

plt.plot(z2(a,z,a,0),x2(z,a,a,0), label='RX (on-axis)')
plt.plot(x,one, label = "RX Tophat (off-axis)")

plt.plot(z2(a,z,a,-np.pi/3),x2(z,a,a,-np.pi/3), label='-pi/3')
plt.plot(z2(a,z,a,np.pi/3),x2(z,a,a,np.pi/3), label='pi/3')
plt.plot(z2(a,z,a,np.pi/6),x2(z,a,a,np.pi/6), label='pi/6')
plt.plot(z2(a,z,a,-np.pi/6),x2(z,a,a,-np.pi/6), label='-pi/6')
plt.plot(z2(a,z,a,np.pi/2),x2(z,a,a,np.pi/2),label='pi/2')
plt.plot(z2(a,z,a,-np.pi),x2(z,a,a,-np.pi),label='pi')


plt.legend()
plt.xlabel('z')
plt.ylabel('x')
plt.grid()
```

## About LO

\begin{equation}
x' = -a+x\cos\alpha + z \sin \alpha = a(\cos\alpha-1) + z \sin \alpha
\end{equation}

\begin{equation}
z' = z\cos\alpha - (x-a) \sin\alpha = z\cos\alpha 
\end{equation}

```python
a =1
one = [a]*11
two = [-a]*11
b = np.pi/4
x_cor=0

def x2(z,x,a,b):
    return(-a+ (x)*np.cos(b)+z*np.sin(b)  )

def z2(x,z,a,b):
    return(z*np.cos(b)-(x)*np.sin(b) )
```
