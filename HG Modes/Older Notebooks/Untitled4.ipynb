```python
import PauLisa as pl, plback as plb,matplotlib.pyplot as plt
```

```python
params = pl.Params(1064e-9,1e-3,0)
plane = pl.Plane(-2e-3,2e-3,200,-2e-3,2e-3,200)
(pl.defaults())
```

```python
modes = plb.modes((0,0,1),(1,1,1),
                 (0,2,1),(2,0,1),(2,1,1),(1,2,1))
```

```python
f = pl.calculate(params,plane,modes,0)
```

```python
fig, ax = plt.subplots(figsize=(12, 12))
cs = plt.contourf(f.plane.getX(), f.plane.getY(), abs(f.getAmp() ** 2))

ax.ticklabel_format(axis='x', style='sci', scilimits=(0, 0), useMathText=True)
ax.ticklabel_format(axis='y', style='sci', scilimits=(0, 0), useMathText=True)

cbar = fig.colorbar(cs)

#plt.xlim([-.5e-4,0])
plt.grid()
```
