```python
import PauLisa as pl
```

```python
modes=pl.Modes((0,1,1))

#can use kwargs any order
params = pl.Params(wavelength=1064e-9,z0=0,w0=0.001)

plane = pl.Plane(xmin=-0.03,xmax=0.03,xpoints=1000,ymin=-0.04,ymax=0.04,ypoints=1000)
print params
```

```python
f1=pl.Calculate(pl.defaultParams,plane,modes,10)
f2=pl.Calculate(pl.defaultParams,plane,modes,50)
```

```python
#single plot
pl.IntensitySliceY(0,f1)

#multiplot
pl.IntensitySliceY(0,f1,f2)

#multiplot over range
pl.IntensitySliceX(0,f1,f2, xlim=[-0.000002,0.000002])

#x scale adjusts to mm and microns
pl.IntensitySliceY(0,f1,f2, xlim=[-.002,0.002])
```

```python
#adjust x and/or y, which also scale to mm/micron
pl.Contour(f1,xlim=[-0.01,0.01],ylim=[-0.01,.01])
pl.Contour(f1,xlim=[-0.002,0.0003])
```

```python
pl.PhaseSliceX(0.1,f1,f2)
```
