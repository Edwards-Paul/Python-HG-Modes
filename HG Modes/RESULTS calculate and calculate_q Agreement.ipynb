# calculate and calculate_q agreement

The Living Reviews article (Bond) gives the general expression for HG modes as

### PauLisa.amplitude
\begin{equation} \label{eq:1}
    u_{nm}(x,y,z)=(2^{n+m-1}n!m!\pi)^{-1/2}
    \frac{1}{w(z)}H_{n} \Big( \frac{\sqrt{2}x}{w(z)} \Big)
    H_{m} \Big(\frac{\sqrt{2}y}{w(z)} \Big)
    \exp \Big(\frac{-ik(x^{2}+y^{2})}{2R_{c}(z)}-
    \frac{x^{2}+y^{2}}{w(z)^{2}} \Big)\;.
\end{equation}

In terms of the Gaussian beam parameter,

\begin{equation} \label{eq:2}
    q(z)=iz_{R}+z-z_{0}=q_{0}+z-z_{0}\;,
\end{equation}

HG modes may be expressed as

### PauLisa.amplitude_q

\begin{align} \label{eq:3}
    u_{nm}(x,y,z)=& u_{n}(x,z)u_{m}(y,z)
    \nonumber\\=& 
    \Big(\frac{2}{\pi}\Big)^{1/4}
    \Big(\frac{1}{2^{n}n!w_{0}}\Big)^{1/2}
    \Big(\frac{q_{0}}{q(z)}\Big)^{1/2}
    \Big(\frac{q_{0}q*(z)}{q_{0}*q(z)}\Big)^{n/2}
    H_{n}
    \exp\Big(\frac{-ik(x^{2})}{2q_{z}}\Big)
    \times u_{m}(y,z)
    \;,
\end{align}

## If each subsequent output matches the previous, then results of calculate and calculate_q are identical..

```python
import PauLisa as pl
```

```python
a = pl.defaultParams
b = pl.defaultPlane
```

## General modes

```python
#calculate and calculate_q comparison


#Calc1=Calc2*1e4
m1=pl.modes((0,0,1))
f1 = pl.calculate(a,b,m1,10)
f2 = pl.calculate_q(a,b,m1,10)
pl.Contour(f1)
pl.Contour(f2)

#Calc1=Calc2*.5e4
m2=pl.modes((0,1,1))
f3a = pl.calculate(a,b,m2,10)
f4a = pl.calculate_q(a,b,m2,10)
pl.Contour(f3a)
pl.Contour(f4a)

#Calc1=Calc2*.5e4
m2=pl.modes((1,0,1))
f3b = pl.calculate(a,b,m2,10)
f4b = pl.calculate_q(a,b,m2,10)
pl.Contour(f3b)
pl.Contour(f4b)

m2=pl.modes((1,1,1))
f3c = pl.calculate(a,b,m2,10)
f4c = pl.calculate_q(a,b,m2,10)
pl.Contour(f3c)
pl.Contour(f4c)

m3=pl.modes((0,1,1),(1,0,1))
f5 = pl.calculate(a,b,m3,10)
f6 = pl.calculate_q(a,b,m3,10)
pl.Contour(f5)
pl.Contour(f6)

m4=pl.modes((2,2,1))
f7 = pl.calculate(a,b,m4,10)
f8 = pl.calculate_q(a,b,m4,10)
pl.Contour(f7)
pl.Contour(f8)


```
