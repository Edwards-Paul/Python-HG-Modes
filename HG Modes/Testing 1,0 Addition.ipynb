## Calculate2 (q parameter)

```python
# spot centers for initial beam and translated beam
# Re[a10]/a00 = delta s/wo , w0 = 1e-3

a10 = .04
a00 = 1



m0 = pl.Modes((0,0,a00))
m1 = pl.Modes((1,0,a10))
m01 = pl.Modes((0,0,a00),(1,0,a10))

f0 =pl.Calculate2(a,b,m0,0)

f1 =pl.Calculate2(a,b,m1,0)
f01 = pl.Calculate2(a,b,m01,0)
f = pl.Calculate2(a,b,m01,0)


pl.IntensitySliceX2(0,f0,f1,f01,labels=['0,0','1,0','Combined'])

print ("predicted x shift: "+str(pl.np.real(a10)/1*a.w0) + " m")
print ("actual: " + str(pl.peakInt(f01).x[0]-pl.peakInt(f0).x[0]) + " m")

print ("% err: " + str(((pl.np.real(a10)/1*a.w0)-(pl.peakInt(f01).x[0]-pl.peakInt(f0).x[0]))*100/(pl.np.real(a10)/1*a.w0)))

print("Combined intensity 0,0 & 1,0 mode, x=y=z=0: " + str(abs(pl.Amplitude(a,0,0,0,m01))**2))
print("Predicted combined intensity, x=y=z=0: " + str(abs(np.sqrt(2/pl.pi)/a.w0)**2))
```

## ADD HG 1,0 mode in phase at z = Zr

```python
# spot centers for initial beam and translated beam
# Re[a10]/a00 = delta s/wo , w0 = 1e-3

a10 = .04
a00 = 1



m0 = pl.Modes((0,0,a00))
m1 = pl.Modes((1,0,a10))
m01 = pl.Modes((0,0,a00),(1,0,a10))

f0 =pl.Calculate2(a,b,m0,a.Zr)
f1 =pl.Calculate2(a,b,m1,a.Zr)
f01 = pl.Calculate2(a,b,m01,a.Zr)
f = pl.Calculate2(a,b,m01,a.Zr)


pl.IntensitySliceX2(0,f0,f1,f01,labels=['0,0','1,0','Combined'])

print ("predicted x shift: "+str(pl.np.real(a10)/1*a.w0))
print ("actual: " + str(pl.peakInt(f01).x[0]-pl.peakInt(f0).x[0]))

print ("% err: " + str(((pl.np.real(a10)/1*a.w0)-(pl.peakInt(f01).x[0]-pl.peakInt(f0).x[0]))*100/(pl.np.real(a10)/1*a.w0)))


phase_adv= pl.Phase(a,0,0,a.Zr,m01)-pl.Phase(a,0,0,0,m01)
#predicted atan(Zr/z)=atan(1)
print ("\nPhase advance:" +str(phase_adv) + "\nPredicted:" +str(pl.atan(1)))
```

# 1mm beam waist, at waist, add HG10 mode in quadrature phase

### 0.04i scale

```python
# Re[a10]/a00 = delta s/wo , w0 = 1e-3

a10 = .04*1j
a00 = 1

m0 = pl.Modes((0,0,a00))
m1 = pl.Modes((1,0,a10))
m = pl.Modes((1,0,1))
m01 = pl.Modes((0,0,a00),(1,0,a10))

f0 =pl.Calculate2(a,c,m0,0)
f1 =pl.Calculate2(a,c,m1,0)
f01 = pl.Calculate2(a,c,m01,0)
fa = pl.Calculate2(a,c,m,0)

pl.IntensitySliceX2(0,f0,f1,f01,labels=['0,0','1,0','Combined'])
pl.PhaseSliceX2(0,f0,f1,f01, labels=['0,0','1,0','Combined'])

theta=a.wavelength/pl.pi/a.w0

pred_theta = theta*abs(a10)



phase_comb=(f01.phase[1,:])
phase_diff=(phase_comb[-1]-phase_comb[0])
phase_ang = phase_diff/(c.getX()[-1]-c.getX()[0])
act_theta = phase_ang/2/np.pi*a.wavelength



print ("\n\nWavefront angle: " + str(act_theta) + " rad" + 
      "\nPredicted angle: " + str(pred_theta) + " rad")
```

### 0.08i scale

```python
# Re[a10]/a00 = delta s/wo , w0 = 1e-3

a10 = .08*1j
a00 = 1

m0 = pl.Modes((0,0,a00))
m1 = pl.Modes((1,0,a10))
m = pl.Modes((1,0,1))
m01 = pl.Modes((0,0,a00),(1,0,a10))

f0 =pl.Calculate(a,c,m0,0)
f1 =pl.Calculate(a,c,m1,0)
f01 = pl.Calculate(a,c,m01,0)
fa = pl.Calculate(a,c,m,0)

pl.IntensitySliceX(0,f0,f1,f01,labels=['0,0','1,0','Combined'])
pl.PhaseSliceX(0,f0,f1,f01, labels=['0,0','1,0','Combined'])

theta=a.wavelength/pl.pi/a.w0

pred_theta = theta*abs(a10)

phase_adv= pl.Phase(a,0,0,a.Zr,m01)-pl.Phase(a,0,0,0,m01)

phase_comb=(f01.phase[1,:])
phase_diff=(phase_comb[-1]-phase_comb[0])
phase_ang = phase_diff/(c.getX()[-1]-c.getX()[0])
act_theta = phase_ang/2/np.pi*a.wavelength

#predicted atan(Zr/z)=atan(1)
print ("Phase advance:" +str(phase_adv) + "\nPredicted:" +str(pl.atan(1)))

print ("\n\nWavefront angle: " + str(act_theta) + " rad" + 
      "\nPredicted angle: " + str(pred_theta) + " rad")

```
