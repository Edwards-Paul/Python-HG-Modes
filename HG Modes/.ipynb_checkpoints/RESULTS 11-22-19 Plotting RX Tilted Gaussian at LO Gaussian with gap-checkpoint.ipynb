# Phase diff of incoming tilted beam on local gaussian for inf. half-plane PD

\begin{equation} \Delta \phi(\alpha) =  \arctan (\frac{\sqrt{2 \pi} w_{0} \alpha}{\lambda})\end{equation}

\begin{align}
\frac{d \Delta \phi}{d \alpha}  
=&
\frac{1}{1+(\frac{\sqrt{2 \pi} w_{0} \alpha}{\lambda})^2}\frac{\sqrt{2 \pi} w_{0}}{\lambda}
\\=&
\frac{\sqrt{2 \pi}w_0 \lambda}{\lambda^2+2 \pi w_0^2 \alpha^2}
\end{align}

Alex's result 

\begin{equation}
\Phi_{dif} = \arctan \left[ \text{erfi}(\frac{kw(z) \sin \alpha}{2 \sqrt{2}}) \right] 
\end{equation}

```python
import PauLisa as pl, numpy as np, matplotlib.pyplot as plt
from scipy.special import erfi as erfi
pi=np.pi
```

### input

```python
#spot, wavelength
w= 1e-3
lam= 1064e-9

#alpha(10nrad)
alpha_min=-700e-6
alpha_max=700e-6
num_points = 1000
alpha = np.linspace(alpha_min,alpha_max,num = num_points)

## PLOT SCALES AND LABELS
micro = 10**(6)
nano = 10**(9)
cycle = 1/(2*pi)
rad = 1
xlab = 'u'
ylab = 'rad'

##SET SCALES
alpha_scale = micro
phase_scale = cycle

if(alpha_scale==nano):
    xlab = 'n'
if(phase_scale==cycle):
    ylab = 'cycles'
```

### phase diff vs. angle

```python
#function    
phase_diff = np.arctan(np.sqrt(2*pi)*(w*alpha)/lam)

#plot (a*1e6 for microrad, diff/2pi for cycles)
plt.figure(figsize=(9,6))
plt.plot(alpha*alpha_scale, phase_diff*phase_scale) 
plt.xlabel(r'Tilt Angle, $\alpha\;$[' + xlab +'rad]', fontsize=15) 
plt.ylabel(r'Phase Diff, $\Delta \phi\;[' + ylab + ']$', fontsize=15) 
plt.title(r'Tilted Gaussian, $\Delta \phi(\alpha) =  \arctan (\frac{\sqrt{2 \pi} w_{0} \alpha}{\lambda})$') 
plt.grid() 
plt.savefig("tilt-initial.png")
```

### Deriv.
$\frac{d \Delta \phi}{d \alpha}$

```python
#function
phase_diff_deriv = (np.sqrt(2*np.pi)*(w*lam))/(lam**2+(2*pi)*((w*alpha)**2))

#plot (a*1e6 for microrad, diff/2pi for cycles)
plt.figure(figsize=(9,6))
plt.plot(alpha*alpha_scale, phase_diff_deriv*phase_scale) 
plt.xlabel(r'Tilt Angle, $\alpha\;$[' + xlab +'rad]',fontsize=15) 
plt.ylabel(r'$ \frac{d \Delta \phi}{d \alpha}$', fontsize=20) 
plt.title(r'Derivative, $\frac{d \Delta \phi}{d \alpha}=\frac{\sqrt{2 \pi}w_0 \lambda}{\lambda^2+2 \pi w_0^2 \alpha^2}$') 
plt.grid()
```

### Computed Deriv. (just checking the previous)
$\frac{d \Delta \phi}{d \alpha}$

```python
#function
comp_phase_diff_deriv = np.diff(phase_diff) / np.diff(alpha)
alpha2 = (alpha[:-1] + alpha[1:]) / 2

#plot (a*1e6 for microrad, diff/2pi for cycles)
plt.figure(figsize=(9,6))
plt.plot(alpha2*alpha_scale, comp_phase_diff_deriv*phase_scale) 
plt.xlabel(r'Tilt Angle, $\alpha\;$[' + xlab +'rad]',fontsize=15) 
plt.ylabel(r'$ \frac{d \Delta \phi}{d \alpha}$', fontsize=20) 
plt.title(r'Comp. Derivative, $\frac{d \Delta \phi}{d \alpha}$') 
plt.grid()
```

### Alex - diff. phase measure
$\Phi_{dif} = \arctan \left[ {erfi}(\frac{kw(z) \sin \alpha}{2 \sqrt{2}}) \right]$

```python
#function
k = 2*pi/lam
diff_phase = np.arctan(erfi(k*w*np.sin(alpha)/(2*np.sqrt(2))))

#plot (a*1e6 for microrad, diff/2pi for cycles)
plt.figure(figsize=(9,6))
plt.plot(alpha*alpha_scale, diff_phase*phase_scale) 
plt.xlabel(r'Tilt Angle, $\alpha\;$[' + xlab +'rad]',fontsize=15) 
plt.ylabel(r'Diff. Phase Measure,$\Phi_{dif}$ [' + ylab + ']', fontsize=15) 
plt.title(r'$\Phi_{dif} = \arctan \left[ {erfi}(\frac{kw(z) \sin \alpha}{2 \sqrt{2}}) \right]$') 
plt.grid()
```

### Comp. deriv

```python
#function
comp_diff_phase_deriv = np.diff(diff_phase) / np.diff(alpha)
alpha2 = (alpha[:-1] + alpha[1:]) / 2

#plot (a*1e6 for microrad, diff/2pi for cycles)
plt.figure(figsize=(9,6))
plt.plot(alpha2*alpha_scale, comp_diff_phase_deriv*phase_scale) 
plt.xlabel(r'Tilt Angle, $\alpha\;$[' + xlab +'rad]',fontsize=15) 
plt.ylabel(r'$ \frac{d \Delta \Phi}{d \alpha}$', fontsize=20) 
plt.title(r'Alex : Diff. Phase Derivative') 
plt.grid()
```
