```python
import PauLisa as pl
```

```python
a=pl.defaultParams
b=pl.Plane(-1e-3,1e-3,40000,-1e-3,1e-3,5)
```

```python
# Re[a10]/a00 = delta s/wo , w0 = 1e-3

a10 = .04
a00 = 1



m0 = pl.Modes((0,0,a00))
m1 = pl.Modes((1,0,a10))
m01 = pl.Modes((0,0,a00),(1,0,a10))

f0 =pl.Calculate2(a,b,m0,6)
f1 =pl.Calculate2(a,b,m1,6)
f01 = pl.Calculate2(a,b,m01,6)

pl.IntensitySliceX2(0,f0,f1,f01)

print ("predicted x shift: "+str(pl.np.real(a10)/1*a.w0))
print ("actual: " + str(pl.peakInt(f01).x-pl.peakInt(f0).x))
```

```python
# Re[a10]/a00 = delta s/wo , w0 = 1e-3

a10 = .04
a00 = 1

m0 = pl.Modes((0,0,a00))
m1 = pl.Modes((1,0,a10))
m01 = pl.Modes((0,0,a00),(1,0,a10))

f0 =pl.Calculate(a,b,m0,2*a.Zr)
f1 =pl.Calculate(a,b,m1,2*a.Zr)
f01 = pl.Calculate(a,b,m01,2*a.Zr)

pl.IntensitySliceX(0,f0,f1,f01)

print ("predicted x shift: "+str(pl.np.real(a10)/1*a.w0))
print ("actual: " + str(pl.peakInt(f01).x-pl.peakInt(f0).x))
print pl.peakInt(f01).loc
print pl.peakInt(f0).loc
```

```python
# Re[a20]/a00 * w0 = (delta w)

a20 = .3
a00 = 1

m0 = pl.Modes((0,0,a00))
m2 = pl.Modes((2,0,a20))
m02 = pl.Modes((0,0,a00),(2,0,a20))

f0 =pl.Calculate(a,b,m0,10)
f2 =pl.Calculate(a,b,m2,10)
f02 = pl.Calculate(a,b,m02,10)

pl.IntensitySliceX(0,f0,f2,f02, xlim=[3.5e-3,4e-3])


print ("predicted "+ str((pl.np.real(a20)/a00)*a.w0))
print ("w at: ", pl.peakInt(f0).peak/pl.e**2,pl.peakInt(f02).peak/pl.e**2)
```
