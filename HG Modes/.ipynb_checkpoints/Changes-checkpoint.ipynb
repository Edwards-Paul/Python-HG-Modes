```python
import PauLisa as pl
```

```python
modes=pl.Modes((0,0,1), (1,0,.1*1j))

pl.ShowModes(modes)
```

```python
f=pl.Calculate(pl.defaultParams,pl.defaultPlane,modes,10)
```

```python
pl.Contour(f)
```

```python
pl.IntensitySliceX(f,0.01)
pl.IntensitySliceY(f,0.01)
```

```python
pl.Phase(pl.defaultParams,0.1,0,100,modes)
```

```python
modes2=pl.Modes((1,0,1),(0,1,.5))
pl.ShowModes(modes2)
f2=pl.Calculate(pl.defaultParams,pl.defaultPlane,modes2,100)
pl.Contour(f2)
pl.IntensitySliceX(f2,0)
pl.IntensitySliceY(f2,0)
```
