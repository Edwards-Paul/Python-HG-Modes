```python
#w(z) and R(z) errors
import PauLisa as pl
import numpy as np
import matplotlib.pyplot as plt
```

```python
params = pl.Params(1064e-9,1e-3,0)

z = np.linspace(0,10e-3,100)

def w(z, params):
    w = params.getW0() * np.sqrt(1 + ((z - params.getZ0()) / params.getZr()) ** 2)
    return w

alpha = 200e-6

z_tilt = z*alpha
```

```python
plt.figure(figsize=(9,6))
plt.plot(z, w(z,params), label='1')
plt.plot(z_tilt, w(z_tilt,params), label='2')
plt.legend()
plt.xlabel(r'z [m]', fontsize=15) 
plt.ylabel(r'w(z) [m]', fontsize=15) 
plt.title(r'w(z)') 
plt.grid()
```
